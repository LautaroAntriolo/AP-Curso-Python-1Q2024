```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

path = "JugadoresDeSelección.xlsx"
jugadores = pd.read_excel(path)
# print(jugadores)
```

```python

promEdad= sum(jugadores["EDAD"])/len(jugadores["EDAD"])
maschicos = jugadores[jugadores['EDAD']<=promEdad]
masgrandes = jugadores[jugadores['EDAD']>=promEdad]

xc = maschicos["NOMBRE"]
yc = maschicos["EDAD"]
xg = masgrandes["NOMBRE"]
yg = masgrandes["EDAD"]

plt.axhline(y=promEdad, color='r', label="promedio")
plt.scatter(xc, yc, label="mas chicos")
plt.scatter(xg,yg, label="mas grandes")
plt.xticks(rotation=90, size=8)
plt.xlabel('Jugadores')
plt.ylabel('Edad')

plt.legend() # Muestro los label que le agregué a cada elemento
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

path = "JugadoresDeSelección.xlsx"
jugadores = pd.read_excel(path)
promEdad= sum(jugadores["EDAD"])/len(jugadores["EDAD"])

plt.figure(figsize=(10,7))
#Grafico puntos que referencien el nombre con la edad de los jugadores
plt.scatter(jugadores["NOMBRE"], jugadores["EDAD"],color="green", label="Edad")

#grafico una linea recta de acuerdo al promedio de edad
plt.axhline(y=promEdad, color='r', label="promedio")

# Grafico una linea punteada desde cada elemento de x hasta su valor de y
plt.vlines(x=jugadores["NOMBRE"], ymin=0, ymax=jugadores["EDAD"], linestyle='--', color='red') 

#roto los label del eje x noventa grados y cambio su font-size de 10 a 8
plt.xticks(rotation=90, size=8)
plt.xlabel('Jugadores')
plt.ylabel('Edad')

plt.legend() # Muestro los label que le agregué a cada elemento
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

fig, ax = plt.subplots()

ax.plot(x, y, 'o-')
ax.vlines(x=2, ymin=0, ymax=4, linestyle='--', color='red')

ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Gráfico de líneas con línea vertical punteada')

plt.show()
```
