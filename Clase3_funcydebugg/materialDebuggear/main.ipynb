Algunos ejemplo para identificar **errores de sintaxis** pueden ser como estos:

```python
# Error de sintaxis
for i in range(10)
    print(i)
```

Saber si una condición se está cumpliendo o no

```python
# Condición que no se cumple
x = 5
if x > 10:
    print("x es mayor que 10")
```

Se puede mejorar con un for:

```python
for i in range(0,20,1):
    if i < 10:
        print("x es mayor que 10")
    else:
        print('ya no se cumple la condición')
```

Ver si un bucle se está ejecutando bien.

```python
# Bucle que no se ejecuta
for i in range(0):
    print(i)
```

```python
# OJO Bucle que no termina
i = 0
while True:
    i += 1
```

Encontrar la linea donde se produce un error: 

```python
# Error en la línea 5
for i in range(10):
    x = 10 / i
    print(x)
```

Encontrar errores en una lista o un diccionario

```python
# Lista fuera de índice
mi_lista = [1, 2, 3]
print(mi_lista[3])
```

```python
# Clave no válida en un diccionario
mi_diccionario = {"a": 1, "b": 2, "c": 3}
print(mi_diccionario["d"])
```
